```python
# 세트 1, 2, 3 전체 데이터 (여기서는 예시로 Set1, Set2를 넣고, 필요시 Set3도 추가 가능)

set1 = [
    ('benefit from', 'Students __________ interactive discussions.', '학생들은 상호작용 토론에서 혜택을 얻는다.', "나는 다양한 경험에서 혜택을 얻었다.", "I benefited from various experiences.", 'Students benefit from interactive discussions.'),
    ('rely on', 'Many people __________ public transportation.', '많은 사람들은 대중교통에 의존한다.', "나는 친구의 조언에 의존했다.", "I relied on my friend\'s advice.", 'Many people rely on public transportation.'),
    ('cope with', 'The company must __________ unexpected challenges.', '그 회사는 예상치 못한 도전에 대처해야 한다.', "나는 스트레스에 잘 대처했다.", "I coped with stress well.", 'The company must cope with unexpected challenges.'),
    ('deal with', 'Managers must __________ conflicts promptly.', '관리자는 갈등을 신속히 처리해야 한다.', "나는 어려운 상황을 처리했다.", "I dealt with a difficult situation.", 'Managers must deal with conflicts promptly.'),
    ('result in', 'Lack of sleep can __________ poor performance.', '수면 부족은 성과 저하를 초래할 수 있다.', "부주의는 사고를 초래했다.", "Carelessness resulted in an accident.", 'Lack of sleep can result in poor performance.'),
    ('lead to', 'Better education can __________ economic growth.', '더 나은 교육은 경제 성장을 이끌 수 있다.', "흡연은 건강 악화를 초래했다.", "Smoking led to poor health.", 'Better education can lead to economic growth.'),
    ('depend on', 'The success of the project __________ teamwork.', '프로젝트의 성공은 팀워크에 달려 있다.', "우리의 계획은 날씨에 달려 있다.", "Our plan depends on the weather.", 'The success of the project depends on teamwork.'),
    ('participate in', 'He decided to __________ the competition.', '그는 대회에 참가하기로 했다.', "나는 봉사활동에 참여했다.", "I participated in volunteer work.", 'He decided to participate in the competition.'),
    ('contribute to', 'Good communication __________ team success.', '좋은 의사소통은 팀 성공에 기여한다.', "그의 아이디어는 프로젝트에 기여했다.", "His idea contributed to the project.", 'Good communication contributes to team success.'),
    ('apply for', 'She will __________ a research grant.', '그녀는 연구 지원금을 신청할 것이다.', "나는 장학금에 지원했다.", "I applied for a scholarship.", 'She will apply for a research grant.'),
    ('apply to', 'These guidelines __________ all employees.', '이 지침은 모든 직원에게 적용된다.', "그 법은 모든 시민에게 적용된다.", "The law applies to all citizens.", 'These guidelines apply to all employees.'),
    ('focus on', 'The lecture __________ climate change.', '강의는 기후 변화에 초점을 맞춘다.', "나는 시험 준비에 집중했다.", "I focused on preparing for the exam.", 'The lecture focuses on climate change.'),
    ('adapt to', 'Immigrants must __________ a new culture.', '이민자들은 새로운 문화에 적응해야 한다.', "나는 새로운 환경에 적응했다.", "I adapted to a new environment.", 'Immigrants must adapt to a new culture.'),
    ('be based on', 'The film is __________ a true story.', '그 영화는 실화를 바탕으로 한다.', "이 보고서는 실제 사례를 기반으로 한다.", "This report is based on real cases.", 'The film is based on a true story.'),
    ('invest in', 'The company __________ renewable energy.', '그 회사는 재생 에너지에 투자한다.', "나는 주식에 투자했다.", "I invested in stocks.", 'The company invests in renewable energy.'),
    ('specialize in', 'The clinic __________ dermatology.', '그 병원은 피부과를 전문으로 한다.', "나는 데이터 분석을 전문으로 한다.", "I specialize in data analysis.", 'The clinic specializes in dermatology.'),
    ('interfere with', 'Loud noise can __________ concentration.', '큰 소음은 집중을 방해할 수 있다.', "휴대폰은 수업을 방해할 수 있다.", "Cell phones can interfere with classes.", 'Loud noise can interfere with concentration.'),
    ('agree with', 'I __________ the proposed changes.', '나는 제안된 변화에 동의한다.', "나는 그의 의견에 동의했다.", "I agreed with his opinion.", 'I agree with the proposed changes.'),
    ('object to', 'Many residents __________ the new policy.', '많은 주민들이 새 정책에 반대한다.', "나는 그 계획에 반대했다.", "I objected to the plan.", 'Many residents object to the new policy.'),
    ('look into', 'The police will __________ the case.', '경찰은 사건을 조사할 것이다.', "나는 문제를 조사했다.", "I looked into the problem.", 'The police will look into the case.'),
    ('account for', 'Exports __________ half of the economy.', '수출은 경제의 절반을 차지한다.', "외국 제품이 판매의 큰 부분을 차지했다.", "Foreign products accounted for a large part of sales.", 'Exports account for half of the economy.'),
    ('believe in', 'Teachers should __________ their students.', '교사는 학생들을 믿어야 한다.', "나는 그의 능력을 믿었다.", "I believed in his ability.", 'Teachers should believe in their students.'),
    ('work on', 'She is __________ her thesis.', '그녀는 논문을 준비 중이다.', "나는 보고서를 작성 중이다.", "I am working on a report.", 'She is working on her thesis.'),
    ('look after', 'Parents __________ their children.', '부모는 자녀를 돌본다.', "나는 조카를 돌봤다.", "I looked after my niece.", 'Parents look after their children.'),
    ('bring up', 'He __________ an interesting topic.', '그는 흥미로운 주제를 꺼냈다.', "나는 회의에서 새로운 아이디어를 제시했다.", "I brought up a new idea at the meeting.", 'He brought up an interesting topic.'),
    ('carry out', 'Scientists __________ experiments.', '과학자들은 실험을 수행한다.', "나는 설문 조사를 진행했다.", "I carried out a survey.", 'Scientists carry out experiments.'),
    ('fill out', 'Please __________ this form.', '이 양식을 작성해 주세요.', "나는 신청서를 작성했다.", "I filled out an application.", 'Please fill out this form.'),
    ('give up', 'Never __________ on your dreams.', '꿈을 절대 포기하지 마라.', "나는 목표를 포기하지 않았다.", "I didn\'t give up on my goal.", 'Never give up on your dreams.')
]

set2 = [
    ("be aware of", "Students should __________ plagiarism rules.", "학생들은 표절 규정을 인식해야 한다.", "나는 위험을 인식했다.", "I was aware of the danger.", "Students should be aware of plagiarism rules."),
    ("be capable of", "She is __________ solving complex problems.", "그녀는 복잡한 문제를 해결할 수 있다.", "나는 문제를 해결할 능력이 있다.", "I am capable of solving problems.", "She is capable of solving complex problems."),
    ("be composed of", "Water is __________ hydrogen and oxygen.", "물은 수소와 산소로 구성되어 있다.", "위원회는 다섯 명으로 구성되어 있다.", "The committee is composed of five members.", "Water is composed of hydrogen and oxygen."),
    ("be concerned about", "Parents are __________ online safety.", "부모는 온라인 안전을 걱정한다.", "나는 결과에 대해 걱정했다.", "I was concerned about the result.", "Parents are concerned about online safety."),
    ("be confident of", "He is __________ his abilities.", "그는 자신의 능력을 확신한다.", "나는 성공을 확신했다.", "I was confident of success.", "He is confident of his abilities."),
    ("be conscious of", "You must be __________ your surroundings.", "주변을 인식해야 한다.", "나는 내 발언을 의식했다.", "I was conscious of my remark.", "You must be conscious of your surroundings."),
    ("be critical of", "The report is __________ the government\'s plan.", "그 보고서는 정부 계획을 비판한다.", "나는 그 정책을 비판했다.", "I was critical of the policy.", "The report is critical of the government\'s plan."),
    ("be curious about", "Children are __________ nature.", "아이들은 자연에 호기심이 많다.", "나는 새 문화에 호기심을 가졌다.", "I was curious about the new culture.", "Children are curious about nature."),
    ("be dependent on", "Some animals are __________ specific habitats.", "일부 동물은 특정 서식지에 의존한다.", "그 나라는 무역에 의존했다.", "The country was dependent on trade.", "Some animals are dependent on specific habitats."),
    ("be devoted to", "She is __________ her research.", "그녀는 연구에 전념한다.", "나는 일에 전념했다.", "I was devoted to my work.", "She is devoted to her research."),
    ("be different from", "This model is __________ the old one.", "이 모델은 이전 것과 다르다.", "내 생각은 그의 생각과 다르다.", "My idea was different from his.", "This model is different from the old one."),
    ("be disappointed with", "They were __________ the results.", "그들은 결과에 실망했다.", "나는 서비스에 실망했다.", "I was disappointed with the service.", "They were disappointed with the results."),
    ("be engaged in", "The students are __________ group work.", "학생들은 그룹 활동에 참여하고 있다.", "나는 프로젝트에 참여했다.", "I was engaged in the project.", "The students are engaged in group work."),
    ("be familiar with", "I am __________ this software.", "나는 이 소프트웨어에 익숙하다.", "나는 그 절차에 익숙했다.", "I was familiar with the procedure.", "I am familiar with this software."),
    ("be famous for", "The city is __________ its cuisine.", "그 도시는 요리로 유명하다.", "그 지역은 와인으로 유명했다.", "The region was famous for its wine.", "The city is famous for its cuisine."),
    ("be filled with", "The room was __________ laughter.", "방은 웃음으로 가득 찼다.", "가방이 책으로 가득 찼다.", "The bag was filled with books.", "The room was filled with laughter."),
    ("be fond of", "She is __________ classical music.", "그녀는 클래식 음악을 좋아한다.", "나는 커피를 좋아했다.", "I was fond of coffee.", "She is fond of classical music."),
    ("be grateful for", "I am __________ your help.", "당신의 도움에 감사한다.", "나는 기회에 감사했다.", "I was grateful for the opportunity.", "I am grateful for your help."),
    ("be guilty of", "He was __________ theft.", "그는 절도죄를 저질렀다.", "나는 실수에 대해 죄책감을 느꼈다.", "I was guilty of a mistake.", "He was guilty of theft."),
    ("be independent of", "The country is __________ foreign influence.", "그 나라는 외국 영향에서 독립했다.", "그 조직은 정부로부터 독립했다.", "The organization was independent of the government.", "The country is independent of foreign influence."),
    ("be interested in", "She is __________ art history.", "그녀는 미술사에 관심이 있다.", "나는 과학에 관심이 가졌다.", "I was interested in science.", "She is interested in art history."),
    ("be involved in", "He is __________ community projects.", "그는 지역사회 프로젝트에 참여한다.", "나는 사건에 연루되었다.", "I was involved in the incident.", "He is involved in community projects."),
    ("be jealous of", "She is __________ her colleague\'s success.", "그녀는 동료의 성공을 질투한다.", "나는 친구의 재능을 질투했다.", "I was jealous of my friend\'s talent.", "She is jealous of her colleague\'s success."),
    ("be known for", "The island is __________ its beaches.", "그 섬은 해변으로 유명하다.", "그는 요리 실력으로 알려졌다.", "He was known for his cooking skills.", "The island is known for its beaches."),
    ("be proud of", "Parents are __________ their children.", "부모는 자녀를 자랑스러워한다.", "나는 내 성취를 자랑스러워했다.", "I was proud of my achievement.", "Parents are proud of their children."),
    ("be related to", "His illness is __________ stress.", "그의 병은 스트레스와 관련이 있다.", "이 문제는 환경과 관련이 있다.", "This issue was related to the environment.", "His illness is related to stress."),
    ("be responsible for", "Drivers are __________ following traffic laws.", "운전자는 교통 법규를 준수할 책임이 있다.", "나는 그 실수에 책임이 있었다.", "I was responsible for the mistake.", "Drivers are responsible for following traffic laws."),
    ("be satisfied with", "I am __________ the outcome.", "나는 결과에 만족한다.", "나는 서비스에 만족했다.", "I was satisfied with the service.", "I am satisfied with the outcome.")
]


set3 = [
    ("approve of", "Many citizens __________ the new policy.", "많은 시민들이 새 정책을 찬성한다.", "나는 그 제안을 찬성했다.", "I approved of the proposal.", "Many citizens approve of the new policy."),
    ("consist of", "The team __________ five members.", "그 팀은 다섯 명으로 구성되어 있다.", "위원회는 세 부서로 구성되었다.", "The committee consisted of three departments.", "The team consists of five members."),
    ("result from", "His illness __________ poor nutrition.", "그의 병은 영양 부족에서 비롯된다.", "문제는 부주의에서 비롯되었다.", "The problem resulted from carelessness.", "His illness results from poor nutrition."),
    ("insist on", "She __________ strict rules.", "그녀는 엄격한 규칙을 고집한다.", "그는 자신의 입장을 고집했다.", "He insisted on his position.", "She insists on strict rules."),
    ("engage with", "Teachers should __________ their students actively.", "교사는 학생들과 적극적으로 교류해야 한다.", "나는 현지 주민들과 교류했다.", "I engaged with local residents.", "Teachers should engage with their students actively."),
    ("be approved for", "The project was __________ government funding.", "그 프로젝트는 정부 지원을 승인받았다.", "내 계획은 승인을 받았다.", "My plan was approved for execution.", "The project was approved for government funding."),
    ("apply oneself to", "He __________ his studies fully.", "그는 학업에 전념한다.", "나는 프로젝트에 전념했다.", "I applied myself to the project.", "He applies himself to his studies fully."),
    ("subscribe to", "Many people __________ online services.", "많은 사람들이 온라인 서비스를 구독한다.", "나는 그 잡지를 구독했다.", "I subscribed to the magazine.", "Many people subscribe to online services."),
    ("account to", "Employees must __________ their manager for progress.", "직원들은 진행 상황을 관리자에게 보고해야 한다.", "나는 팀장에게 결과를 보고했다.", "I accounted to my team leader.", "Employees must account to their manager for progress."),
    ("count on", "You can always __________ me for help.", "너는 항상 나에게 도움을 기대할 수 있다.", "나는 친구에게 도움을 기대했다.", "I counted on my friend for help.", "You can always count on me for help."),
    ("cut down on", "We need to __________ expenses this year.", "우리는 올해 지출을 줄여야 한다.", "나는 설탕 섭취를 줄였다.", "I cut down on sugar.", "We need to cut down on expenses this year."),
    ("deal in", "The shop __________ rare books.", "그 가게는 희귀 서적을 취급한다.", "그 상인은 보석을 취급했다.", "The merchant dealt in jewelry.", "The shop deals in rare books."),
    ("depend upon", "Our success __________ market conditions.", "우리의 성공은 시장 상황에 달려 있다.", "그 계획은 자원에 달려 있었다.", "The plan depended upon resources.", "Our success depends upon market conditions."),
    ("fall back on", "He had to __________ his savings.", "그는 저축에 의존해야 했다.", "나는 친구의 도움에 의존했다.", "I fell back on my friend\'s help.", "He had to fall back on his savings."),
    ("go through", "We must __________ the report before publishing.", "우리는 출판 전에 보고서를 검토해야 한다.", "나는 힘든 시기를 겪었다.", "I went through a difficult time.", "We must go through the report before publishing."),
    ("interfere in", "They should not __________ other countries\' affairs.", "그들은 다른 나라의 일에 간섭해서는 안 된다.", "그는 내 일에 간섭했다.", "He interfered in my business.", "They should not interfere in other countries\' affairs."),
    ("keep up with", "I can\'t __________ the latest news.", "나는 최신 소식을 따라잡을 수 없다.", "나는 수업을 따라갔다.", "I kept up with the class.", "I can\'t keep up with the latest news."),
    ("look forward to", "I __________ meeting you.", "나는 당신을 만나기를 고대한다.", "나는 여행을 고대했다.", "I looked forward to the trip.", "I look forward to meeting you."),
    ("make up for", "This refund will __________ the inconvenience.", "이 환불은 불편을 보상할 것이다.", "나는 실수를 만회했다.", "I made up for my mistake.", "This refund will make up for the inconvenience."),
    ("pay attention to", "Please __________ the instructions carefully.", "지시를 주의 깊게 따라 주세요.", "나는 세부 사항에 주의를 기울였다.", "I paid attention to the details.", "Please pay attention to the instructions carefully."),
    ("provide for", "Parents must __________ their children\'s needs.", "부모는 자녀의 필요를 제공해야 한다.", "나는 가족을 부양했다.", "I provided for my family.", "Parents must provide for their children\'s needs."),
    ("run out of", "We have __________ time.", "우리는 시간이 다 떨어졌다.", "나는 돈이 떨어졌다.", "I ran out of money.", "We have run out of time."),
    ("set up", "They will __________ a new company.", "그들은 새 회사를 설립할 것이다.", "나는 전시회를 준비했다.", "I set up the exhibition.", "They will set up a new company."),
    ("stick to", "You must __________ the schedule.", "너는 일정을 지켜야 한다.", "나는 계획을 고수했다.", "I stuck to my plan.", "You must stick to the schedule."),
    ("take part in", "Students will __________ the workshop.", "학생들은 워크숍에 참여할 것이다.", "나는 회의에 참석했다.", "I took part in the meeting.", "Students will take part in the workshop."),
    ("talk over", "Let\'s __________ the details later.", "세부 사항을 나중에 논의하자.", "우리는 문제를 상의했다.", "We talked over the issue.", "Let\'s talk over the details later."),
    ("watch out for", "You should __________ pickpockets.", "소매치기를 조심해야 한다.", "나는 길 건너는 사람을 조심했다.", "I watched out for pedestrians.", "You should watch out for pickpockets."),
    ("work out", "They will __________ a solution soon.", "그들은 곧 해결책을 마련할 것이다.", "나는 문제를 해결했다.", "I worked out the problem.", "They will work out a solution soon.")
]


# set1, set2, set3를 한 PDF에 묶되, 각 세트별로 문제/답 순서가 따로 나오게 하고, 문제별로 페이지가 넘어가지 않게 처리

from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, PageBreak, KeepTogether
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics
import os

# 폰트 등록
font_path = os.path.join(os.getcwd(), 'malgun.ttf')
pdfmetrics.registerFont(TTFont('malgun.ttf', font_path))

styles = getSampleStyleSheet()
style_normal = ParagraphStyle(
    'Normal',
    parent=styles['Normal'],
    fontName='malgun.ttf',
    fontSize=12,
    leading=15,
    spaceBefore=0,
    spaceAfter=0,
)
style_heading = ParagraphStyle(
    'Heading1',
    parent=styles['Heading1'],
    fontName='malgun.ttf',
    fontSize=9,  # 작게
    spaceAfter=3,
    spaceBefore=0,
    leftIndent=0,
    rightIndent=0,
    alignment=0,  # 왼쪽 정렬
)

answer_style = ParagraphStyle(
    'Answer',
    parent=styles['Normal'],
    fontName='malgun.ttf',
    fontSize=9,  # 더 작게
    leading=11,
    spaceBefore=0,
    spaceAfter=0,
)

def add_set_to_content(set_data, set_name, content):
    # 문제지
    content.append(Paragraph(f"{set_name} 문제지", style_heading))
    for page_start in range(0, len(set_data), 7):
        page_items = set_data[page_start:page_start+7]
        page_content = []
        for i, (expr, example_blank, example_ko, writing_ko, writing_en, example_answer) in enumerate(page_items, start=page_start+1):
            problem = [
                Paragraph(f"{i}. __________", style_normal),
                Paragraph(f"예문: {example_blank}", style_normal),
                Paragraph(f"해석: {example_ko}", style_normal),
                Paragraph(f"작문: {writing_ko}", style_normal),
                Spacer(1, 8),
                Paragraph("_________________________________________________", style_normal),
                Spacer(1, 11)
            ]
            page_content.append(KeepTogether(problem))
        content.extend(page_content)
        # 7개씩 한 페이지에, 마지막 페이지는 PageBreak 추가하지 않음
        if page_start + 7 < len(set_data):
            content.append(PageBreak())

    # 답지 (별도의 페이지에)
    content.append(PageBreak())
    content.append(Paragraph(f"{set_name} 답지", style_heading))
    answer_style = ParagraphStyle(
        'Answer',
        parent=styles['Normal'],
        fontName='malgun.ttf',
        fontSize=9,  # 더 작게
        leading=11,
        spaceBefore=0,
        spaceAfter=0,
    )
    for i, (expr, example_blank, example_ko, writing_ko, writing_en, example_answer) in enumerate(set_data, start=1):
        answer_text = f"{i}. {expr} — {example_answer} / {writing_en}"
        content.append(Paragraph(answer_text, answer_style))
        content.append(Spacer(1, 4))
    content.append(PageBreak())

content = []
add_set_to_content(set1, "Set1", content)
add_set_to_content(set2, "Set2", content)
add_set_to_content(set3, "Set3", content)

pdf_path = 'TOEFL_all_sets_blank_answerkey.pdf'
doc = SimpleDocTemplate(pdf_path)
doc.build(content)

pdf_path
```
